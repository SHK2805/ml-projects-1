```python
import pandas as pd
```

```python
# import dataset
df = pd.read_csv("/content/Credit_default_dataset.csv")
```

```python
# get size
df.size
```

```python
# print data
df.head(5)
```

```python
# Remove ID column
# axis = 1 refers to columns
df.drop("ID", axis=1, inplace=True)
```

```python
# print data
df.head(5)
```

```python
# rename the column PAY_0 to PAY_1 to make the column names even
df.rename(columns={"PAY_0":"PAY_1"}, inplace=True)
```

```python
# print data
df.head(5)
```

```python
# see counts of column sex
df["SEX"].value_counts()
```

```python
# see counts of column EDUCATION
df["EDUCATION"].value_counts()
```

```python
# map 0, 5, 6 to 4
# in the original dataset there are no values for catogories 0, 5, 6
# so we assumed that they are higher education that can be mapped to 4
df["EDUCATION"] = df["EDUCATION"].map({0:4, 1:1, 2:2, 3:3, 4:4, 5:4, 6:4})
```

```python
df["MARRIAGE"].value_counts()
```

```python
# mapping 0 to 3 for MARRIAGE
df["MARRIAGE"] = df["MARRIAGE"].map({0:3, 1:1, 2:2, 3:3})
```

```python
df.head(5)
```

```python
# get the standard scalar
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
# split dependent and independent features
X = df.drop("default.payment.next.month", axis=1)
y = df["default.payment.next.month"]
```

```python
# scale X
X = scalar.fit_transform(X)
```

```python
# xgboost hyper-parameters
params = {
    "learning_rate": [0.005, 0.01],
    "max_depth":[4,5,6],
    "min_child_weight":[1,3,5],
    "gamma": [0.00, 0.01],
    "colsample_bytree": [0.5, 0.7]
}
```

```python
# randomized search cv
from sklearn.model_selection import RandomizedSearchCV
import xgboost as xgb
classifier = xgb.XGBClassifier()
random_search = RandomizedSearchCV(classifier,
                                   param_distributions=params,
                                   n_iter=5,
                                   scoring="roc_auc",
                                   n_jobs=-1,
                                   cv=5,
                                   verbose=3)
```

```python
# fit
random_search.fit(X, y)
```

```python
# best params
random_search.best_params_
```

```python
# best indicator
random_search.best_estimator_
```

```python
best_classifier = xgb.XGBClassifier(base_score=None, booster=None, callbacks=None,
              colsample_bylevel=None, colsample_bynode=None,
              colsample_bytree=0.7, device=None, early_stopping_rounds=None,
              enable_categorical=False, eval_metric=None, feature_types=None,
              gamma=0.01, grow_policy=None, importance_type=None,
              interaction_constraints=None, learning_rate=0.005, max_bin=None,
              max_cat_threshold=None, max_cat_to_onehot=None,
              max_delta_step=None, max_depth=6, max_leaves=None,
              min_child_weight=3, monotone_constraints=None,
              multi_strategy=None, n_estimators=None, n_jobs=None,
              num_parallel_tree=None, random_state=None)
```

```python
# cross validation
from sklearn.model_selection import cross_val_score
score = cross_val_score(best_classifier, X, y, cv=10)
score
```
